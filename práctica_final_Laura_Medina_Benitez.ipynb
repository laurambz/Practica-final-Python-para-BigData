```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
```

```python
# A. Una función que calcule el precio total de la compra

from functools import reduce
def compra_total(articulos, compra):
    carrito =map(lambda x: articulos[x], compra)
    return reduce(lambda x, y: x + y, carrito, 0)
print(compra_total(articulos,compra))
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%

art_desc = dict(map(lambda x: (x[0], x[1]*0.9), articulos.items()))

def compra_total_desc(art_desc, compra):
    carrito =map(lambda x: art_desc[x], compra)
    return reduce(lambda x, y: x + y, carrito, 0)
print(compra_total(art_desc,compra))
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def descuento(articulos,compra):
    if compra_total(articulos,compra)>100:
        return compra_total_desc(articulos,compra)

print(f"Mi compra total es de {descuento(articulos,compra)} €")
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)

def compra_IVA(articulos,compra):
    return descuento(articulos,compra) * 1.21

print(f"Mi compra total con IVA incluido es de {compra_IVA(articulos,compra)} €")   
```
